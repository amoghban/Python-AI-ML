```python
myList = [10, 15, 19, 24, 41, 42, 46, 51, 60]

position = [3,4,6] 
newList = []
index = 0
for i in myList:
    if index in position:
        print(f"Element at index {index}: {i}")
        newList.append(i)
    index +=1
print("#######")
# Another approach
for j in position:
    print(f"Element at index {j}: {myList[j]}")

print(newList)
```

Q9: Replace index position 1,4,6 with 25,35,45

```python
myList = [10, 15, 19, 24, 41, 42, 46, 51, 60]
replacements = {1: 25, 4: 35, 6: 45}

print(f"Initial List:  {myList}")
# Way 1
for index , val in replacements.items():
    myList[index] = val

#Way 2
for index in replacements.keys():
    myList[index] = replacements.get(index)


print(f"Modified list: {myList}")
```

Q16 Sum of 2 lists at the same index.

```python
myList1 = [1,2,3]
myList2 = [4,5,6]

newList = []
indexi = 0
for i in myList1:
    indexj = 0
    for j in myList2:
        if indexi == indexj:
            newList.append(i+j)
            break
        indexj += 1
    indexi += 1

print(f"Added List : {newList}")


# Another solution

myList1 = [1,2,3]
myList2 = [4,5,6]
newList = []
for i in range(len(myList1)):
    newList.append(myList1[i] + myList2[i])

print(f"Added List : {newList}")
```
