Q1. Write a program that uses a while loop to calculate the sum of the first 50 natural numbers?

```python
x=1
sum=0
while x<=50:
    sum=sum+x
    x=x+1
print(f"sum of the first 50 natural numbers: {sum}")
```

Q16. create an empty list. Continuously ask the user t o enter their favourite movies until they type "done." Use the append() method to add each movie to the list?

```python
print("To Exit Loop enter done.")
userInput = input("Enter your favourite movie name: ")

favMovies = []
while userInput.upper() != "DONE":
    favMovies.append(userInput)
    userInput = input("Enter your favourite movie name: ")
print(f"Entered List of favourite Movies: {favMovies}")
```

Q17. write a program to accept 10 numbers from the user and display the largest and smallest number?

```python
x=1
smallest = largest = 0

while x<= 10:
    inputNumber = eval(input("Enter number {x}: "))
    if inputNumber < smallest or smallest==0:
        smallest = inputNumber
    elif inputNumber > largest:
        largest = inputNumber
    x +=1
print(f"Largest: {largest} \n smallest: {smallest}")
```

Q24. Write a program using a while loop to print a right-angled triangle of asterisks (*)?

```python
# Ask the user for the height of the triangle
height = int(input("Enter the height of the triangle: "))

# Initialize the starting row
row = 1

# Use a while loop to print each row
while row <= height:
    print("*" * row)
    row += 1
```

Q26. Write a program that uses a while loop to find the average of positive numbers entered by the user until they enter a negative number?

```python
inputNumber,average = 0,0
while inputNumber >= 0:
    inputNumber = eval(input("Enter a number (Negative number to Exit): "))
    if average ==0:
        average = inputNumber
    else:    
        average = (average + inputNumber)/2
    print(f"Average: {average}")
```
